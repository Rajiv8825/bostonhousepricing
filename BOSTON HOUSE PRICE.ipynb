```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
df=load_boston()
```

```python
df.keys()
```

```python
df
```

```python
dataset=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
dataset.head()
```

```python
dataset["Price"]=df.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.corr()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(dataset.corr(),annot=True, cmap=plt.cm.CMRmap_r)
```

```python
discrete_feature=[feature for feature in dataset.columns if dataset[feature].nunique()<25]
```

```python
dataset[discrete_feature].nunique()
```

```python
for feature in discrete_feature:
    
    sns.barplot(data=dataset,x=feature, y=dataset["Price"])
    plt.show()
```

```python
for feature in dataset.columns:
    sns.regplot(data=dataset, x=feature, y=dataset["Price"])
    plt.show()
```

```python
# model training
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=42) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
reg_pred=regression.predict(x_test)
```

```python
sns.scatterplot(x=y_test, y=reg_pred)
plt.xlabel("y_test")
plt.ylabel("Prediction")
plt.show()
```

```python
residual=y_test-reg_pred
```

```python
residual
```

```python
sns.displot(residual,kind="kde")
```

```python
#it is follwing gaussian distribution but there are some outliers after 10(10-30)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(r2_score(y_test,reg_pred))
```
